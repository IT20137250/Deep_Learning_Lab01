```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create random arrays
exponential_array = np.random.exponential(scale=1, size=100000)
uniform_array = np.random.rand(100000)
normal_array = np.random.normal(size=100000)

# Plot histograms
plt.hist(exponential_array, density=True, bins=100, histtype="step", color="blue", label="Exponential")
plt.hist(uniform_array, density=True, bins=100, histtype="step", color="green", label="Uniform")
plt.hist(normal_array, density=True, bins=100, histtype="step", color="red", label="Normal")
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Create a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate Z = X**2 + Y**2
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Z = X**2 + Y**2')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
# Replace 'path_to_dataset.csv' with the actual path to your dataset
df = pd.read_csv('path_to_dataset.csv')

# Select the desired features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Calculate Pearson correlation matrix
pearson_corr = df[features].corr()

# Calculate Spearman correlation matrix
spearman_corr = df[features].corr(method='spearman')

# Create subplots for Pearson and Spearman correlation heatmaps
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot Pearson correlation heatmap
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', center=0, ax=ax1)
ax1.set_title('Pearson Correlation')

# Plot Spearman correlation heatmap
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', center=0, ax=ax2)
ax2.set_title('Spearman Correlation')

plt.tight_layout()
plt.show()
```
